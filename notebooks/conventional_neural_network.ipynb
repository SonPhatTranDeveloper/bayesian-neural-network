# Demonstration of Conventional Neural Network implementation

```python
import matplotlib.pyplot as plt
from datasets import generate_sinusoidal
from network.conventional_neural_network import ConventionalNeuralNetwork
import numpy as np
```

## Generate synthetic sinusoidal dataset

```python
# Generate the dataset
x, t = generate_sinusoidal(low=-10, high=10, size=100, scale=0.1)
```

```python
# Plot the sinusoidal dataset
plt.scatter(x, t)
plt.xlabel('x')
plt.ylabel('t')
plt.title('Sinusoidal data')
plt.show()
```

## Fit a neural network and visualize its prediction

```python
# Create and train neural network
nn = ConventionalNeuralNetwork(input_size=1, hidden_size=500)

# Define the hyperparameters
EPOCHS = 2000
LEARNING_RATE = 0.0001

# Run the training loop
for epoch in range(EPOCHS):
    # Get the error and output
    output, e = nn.forward(x, t)
    
    # Display the error
    print(f"EPOCHS {epoch}, Score {e}")
    
    # Perform back-propagation
    nn.backward(learning_rate=LEARNING_RATE)
```

## Visualize the prediction

```python
# Generate the prediction
x_test = np.arange(-10, 10, 0.01).reshape(-1, 1)
y_test = nn.predict(x_test)
```

```python
# Plot the predictions and the true datapoint
plt.plot(x_test, y_test, color='blue')
plt.scatter(x, t, color='orange')
plt.xlabel('x')
plt.ylabel('t')
plt.title('Sinusoidal data prediction')
plt.show()
```
